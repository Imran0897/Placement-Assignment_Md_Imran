```python
import pandas as pd
import numpy as np
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances, manhattan_distances,haversine_distances
```

```python
df = df.drop(['link','authors', 'date'],axis=1)
```

```python
df.head()
```

```python
df['text'] = df['headline'] + ' ' + df['short_description']
```

```python
df.head(2)
```

```python
# Vectorize the text data
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['text'])
```


1. Cosine Similarity Algorithm

```python
# Function to find the most similar data using different similarity algorithm
def find_similar_data_cosine(query, top_n=5):

    # Vectorize the query
    query_vector = vectorizer.transform([query])

    # Calculate similarities using different algorithms
    cosine_sim = cosine_similarity(X, query_vector).flatten()

    # Find the indices of top similar data points
    top_indices = cosine_sim.argsort()[-top_n:][::-1]

    # Return the top similar data points
    similar_data = df.iloc[top_indices]

    return similar_data
```

```python
# Example usage
query = "parents nowadays"
similar_data_cosine = find_similar_data_cosine(query)
```

2. Euclidean Similarity Algorithm

```python
# Function to find the most similar data using different similarity algorithm
def find_similar_data_euclidean(query, top_n=5):
    # Vectorize the query
    query_vector = vectorizer.transform([query])

    euclidean_sim = euclidean_distances(X, query_vector).flatten()

    # Find the indices of top similar data points
    top_indices = euclidean_sim.argsort()[-top_n:][::-1]

    # Return the top similar data points
    similar_data = df.iloc[top_indices]

    return similar_data
```

```python
# Example usage
query = "parents nowadays"
similar_data_euclidean = find_similar_data_euclidean(query)
```

```python
# Function to find the most similar data using different similarity algorithm
def find_similar_data_manhattan(query, top_n=5):
    # Vectorize the query
    query_vector = vectorizer.transform([query])

    manhattan_sim = manhattan_distances(X, query_vector).flatten()

    # Find the indices of top similar data points
    top_indices = manhattan_sim.argsort()[-top_n:][::-1]

    # Return the top similar data points
    similar_data = df.iloc[top_indices]

    return similar_data
```
