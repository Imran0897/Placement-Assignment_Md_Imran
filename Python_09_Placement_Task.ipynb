```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv("https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD")
```

```python
df.head()
```

● Get all the cars and their types that do not qualify for clean alternative fuel vehicle

```python
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
```

```python
# cars which didn't qualify for clean alternative feul vehicle
df_1 = df[(df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] != 'Clean Alternative Fuel Vehicle Eligible')]
```

```python
df_not_qualify = df_1[['Model','Electric Vehicle Type','Clean Alternative Fuel Vehicle (CAFV) Eligibility']]
```

```python
df_not_qualify
```

```python
fig, ax = plt.subplots(figsize=(16, 5))
sns.countplot(x='Clean Alternative Fuel Vehicle (CAFV) Eligibility', data=df_not_qualify)

plt.show()
```

Get all TESLA cars with the model year, and model type made in Bothell City.

```python
# All Tesla cars in Brothell city
df_tesla = df[(df['Make']=='TESLA') & (df['City']=='Bothell')]
```

```python
df_tesla.head()
```

```python
df_tesla_details = df_tesla[['Model Year','Electric Vehicle Type']]
```

```python
# All Tesla cars with year and model type in Brothell city
df_tesla_details
```

```python
plt.figure(figsize=(10, 6))
df_tesla_details['Model Year'].plot(kind='hist', bins=20)
plt.xlabel("year")
plt.ylabel("Number of car")
plt.title("Tesla car in Bothell city")
plt.show()
```

Get all the cars that have an electric range of more than 100, and were made after
2015

```python
df.head(1)
```

```python
# all the cars that have an electric range of more than 100, and were made after 2015
df_tesla_range = df[(df['Electric Range']>100) & (df['Model Year']>2015)]
```

```python
df_tesla_range.head()
```

```python
df_tesla_range_details = df_tesla_range[['Make','Model']]
```

```python
df_tesla_range_details
```

```python
fig, ax = plt.subplots(figsize=(16, 5))
sns.countplot(x='Make', data=df_tesla_range_details)
plt.show()
```

Draw plots to show the distribution between city and electric vehicle type

```python
# type and number electric vehicle in each city
df.groupby('City')['Electric Vehicle Type'].value_counts().unstack()
```

```python
plt.figure(figsize=(30, 25))
df.groupby('City')['Electric Vehicle Type'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.title('Distribution of Electric Vehicle Types by City')
plt.xlabel('City')
plt.ylabel('Count')
plt.legend(title='Electric Vehicle Type')
plt.show()
```
