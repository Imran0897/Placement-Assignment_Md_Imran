```python
df['Weekend'] = df['Weekend'].map({False:0, True:1})
df['Revenue'] = df['Revenue'].map({False:0, True:1})
```

Handling categorical data

```python
col_cat = []
```

```python
for i in [i for i in df.columns if df[i].dtype == 'object']:
  col_cat.append(i)
```

```python
col_cat
```

```python
df_month = pd.get_dummies(df[['Month']])
```

```python
df_VisitorType = pd.get_dummies(df[['VisitorType']])
```

```python
df_VisitorType
```

```python
df.drop(columns=col_cat,inplace=True)
```

```python
df_final =pd.concat([df_month,df_VisitorType,df],axis=1)
```

```python
df_final
```

```python
y = df[['Weekend','Revenue']]
```

```python
X = df.drop(columns = ['Weekend','Revenue'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Model Building

```python
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from xgboost import XGBClassifier,XGBRegressor
from sklearn.metrics import accuracy_score,classification_report,r2_score
from sklearn.linear_model import LogisticRegression
```

```python
model_rdf = RandomForestClassifier()

model_rdf.fit(X_train,y_train)
```

```python
y_pred_rdf = model_rdf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_rdf)
```

```python
print('Classification Report of Random Forest Classifier : \n',classification_report(y_test,y_pred_rdf, digits=4))
```

```python
model_xgb = XGBClassifier()

model_xgb.fit(X_train,y_train)
```

```python
y_pred_xgb = model_xgb.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_xgb)
```

```python
print('Classification Report of Random Forest Classifier : \n',classification_report(y_test,y_pred_xgb, digits=4))
```
