```python
import requests
import pandas as pd
#loading data from url
response = requests.get("https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json")
data = response.json()

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data['pokemon'])
```
