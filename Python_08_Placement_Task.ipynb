```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Download the data
response = requests.get("http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes")
data = response.json()
```

```python
data['_embedded']
```

```python
data['_embedded']['episodes']
```

```python
# Convert the data to a pandas DataFrame
df = pd.DataFrame(data["_embedded"]['episodes'])
```

```python
df.head()
```

```python
df.info()
```

1. Get all the overall ratings for each season and using plots compare the ratings for all the
seasons, like season 1 ratings, season 2, and so on.

```python
df['average_rating'] = df['rating'].apply(lambda x:x['average'])
```

```python
df_new1 = df[['season','average_rating']].groupby('season')
```

```python
df_2 = df_new1.mean()
```

```python
plt.figure(figsize=(25, 20))
df_2.plot(kind='bar')
plt.xlabel("season")
plt.ylabel("average rating")
plt.title("Average rating season wise")
plt.show()
```

Get all the episode names, whose average rating is more than 8 for every season

```python
df_3 = df[df['average_rating']>8][['season','number','name','average_rating']]
```

```python
df_3
```

```python
plt.figure(figsize=(10, 10))
df_3.plot(x='name', y='average_rating',kind='bar')
plt.xlabel("epsiode name")
plt.ylabel("average rating")
plt.title("episode with average rating more than 8")
plt.show()
```
